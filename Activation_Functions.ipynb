# Activation functions in Neural Networks

### 1. Sigmoid Function

```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Inputs
inputs = np.array([13, 3, -12])

# Compute sigmoid outputs for the inputs
outputs = sigmoid(inputs)

# Print the results
for x in inputs:
    print('{} --> {}'.format(x, sigmoid(x)))

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Create a range of input values
x_values = np.linspace(-10, 10, 400)
sigmoid_values = sigmoid(x_values)

# Plotting
plt.plot(x_values, sigmoid_values, label='Sigmoid', color='blue')
plt.axhline(0.5, color='black', lw=0.5, ls='--')  # Add horizontal line at y=0.5
plt.axvline(0, color='black', lw=0.5, ls='--')    # Add vertical line at x=0
plt.title('Sigmoid Activation Function')
plt.xlabel('Input Value')
plt.ylabel('Sigmoid Output')
plt.ylim(-0.1, 1.1)
plt.xlim(-10, 10)
plt.grid()
plt.legend()
plt.show()
```

### 2. Tanh Function

```python
import numpy as np

# Tanh activation function
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# Inputs
inputs = np.array([13, 3, -12])

# Compute tanh outputs for the inputs
outputs = tanh(inputs)

# Print the results
for x in inputs:
    print('{} --> {}'.format(x, tanh(x)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh activation function
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# Create a range of input values
x_values = np.linspace(-10, 10, 400)
tanh_values = tanh(x_values)

# Plotting
plt.plot(x_values, tanh_values, label='tanh(x)', color='blue')  # Plot the Tanh function
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Tanh Function Output")
plt.axhline(0, color='black', lw=0.5, ls='--')  # Add horizontal line at y=0
plt.axvline(0, color='black', lw=0.5, ls='--')  # Add vertical line at x=0
plt.grid()
plt.legend()
plt.show()

```

### 3. ReLU Function

```python
import numpy as np

# ReLU activation function
def relu(x):
    return np.maximum(0.0, x)

# Inputs
inputs = np.array([13, 3, -12])

# Compute ReLU outputs for the inputs
outputs = relu(inputs)

# Print the results
for x in inputs:
    print('{} --> {}'.format(x, relu(x)))

```

```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU activation function
def relu(x):
    return np.maximum(0.0, x)

# Create a range of input values
x_values = np.linspace(-10, 10, 400)
relu_values = relu(x_values)

# Plotting
plt.plot(x_values, relu_values, label='ReLU(x)', color='blue')  # Plot the ReLU function
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.title("ReLU Function Output")
plt.axhline(0, color='black', lw=0.5, ls='--')  # Add horizontal line at y=0
plt.axvline(0, color='black', lw=0.5, ls='--')  # Add vertical line at x=0
plt.grid()
plt.legend()
plt.show()
```

### 4.Leaky ReLU Function

```python
import numpy as np

# Leaky ReLU activation function
def leakyrRelu(x):
    return np.maximum(0.1*x,x)

# Inputs
inputs = np.array([13, 3, -12])

# Compute ReLU outputs for the inputs
outputs = leakyrRelu(inputs)

# Print the results
for x in inputs:
    print('{} --> {}'.format(x, leakyrRelu(x)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky ReLU activation function
def leakyrRelu(x):
    return np.maximum(0.1 * x, x)

# Create a range of input values
x_values = np.linspace(-10, 10, 400)
leaky_relu_values = leakyrRelu(x_values)

# Plotting
plt.plot(x_values, leaky_relu_values, label='Leaky ReLU(x)', color='blue')  # Plot the Leaky ReLU function
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.title("Leaky ReLU Function Output")
plt.axhline(0, color='black', lw=0.5, ls='--')  # Add horizontal line at y=0
plt.axvline(0, color='black', lw=0.5, ls='--')  # Add vertical line at x=0
plt.grid()
plt.legend()
plt.show()
```
